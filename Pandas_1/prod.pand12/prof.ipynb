```python
import pandas as pd
melb_df = pd.read_csv('data/melb_data_fe.csv')
```

```python
melb_df['Date'] = pd.to_datetime(melb_df['Date'], dayfirst= False)
melb_df['Date'].dt.quarter.value_counts()
```

```python
cols_to_exclude = ['Date', 'Rooms', 'Bedroom', 'Bathroom', 'Car'] 
max_unique_count = 150 
for col in melb_df.columns:
    if melb_df[col].nunique() < max_unique_count and col not in cols_to_exclude:
        melb_df[col] = melb_df[col].astype('category') 
```

```python
mask1 = melb_df['AreaRatio'] < -0.8
mask2 = melb_df['Type'] == 'townhouse'
mask3 = melb_df['SellerG'] == 'McGrath'
melb_df[mask1 & mask2 & mask3].sort_values(
    by=['Date', 'AreaRatio'],
    ascending=[True, False],
    ignore_index=True
).loc[:, ['Date', 'AreaRatio']]
```

```python
melb_df = melb_df.sort_values(by='AreaRatio', ascending=False, ignore_index = True)
```

```python
heh = melb_df[(melb_df['Type'] == 'townhouse') & (melb_df['Rooms'] > 2)]
heh = heh.sort_values(by = ['Rooms', 'MeanRoomsSquare'], ascending = [True, False], ignore_index = True)
heh['Price'].loc[18]
```
